# 📊 SentinelAI - Dataset Analysis

This notebook analyzes the synthetic dataset used for AI-based proctoring.

### Objectives:
- Understand dataset structure
- Analyze class distribution
- Visualize sample images
- Validate dataset integrity

```python
import os
import json
import random
import cv2
import matplotlib.pyplot as plt
from collections import Counter

plt.rcParams["figure.figsize"] = (10, 5)
```

```python
# Project root assumption
BASE_PATH = "../../"

DATASET_PATH = os.path.join(BASE_PATH, "dataset")
IMAGE_PATH = os.path.join(DATASET_PATH, "images")
VIDEO_PATH = os.path.join(DATASET_PATH, "videos")
AUDIO_PATH = os.path.join(DATASET_PATH, "audio")
ANNOTATION_PATH = os.path.join(DATASET_PATH, "annotations")

LABELS_FILE = os.path.join(ANNOTATION_PATH, "labels.json")

print("Dataset Path:", DATASET_PATH)
```

```python
required_dirs = [
    IMAGE_PATH,
    VIDEO_PATH,
    AUDIO_PATH,
    ANNOTATION_PATH
]

for d in required_dirs:
    print(f"{d} ->", "✅ Exists" if os.path.exists(d) else "❌ Missing")
```

```python
with open(LABELS_FILE, "r") as f:
    annotations = json.load(f)

print("Total annotated samples:", len(annotations))

# Show first record
annotations[0]
```

```python
labels = [item["label"] for item in annotations]
label_counts = Counter(labels)

print("\nClass Distribution:")
for k, v in label_counts.items():
    print(f"{k}: {v}")
```

```python
plt.bar(label_counts.keys(), label_counts.values())
plt.xticks(rotation=45)
plt.title("Class Distribution")
plt.xlabel("Classes")
plt.ylabel("Count")
plt.show()
```

```python
missing_files = []

for item in annotations:
    if not os.path.exists(item["file"]):
        missing_files.append(item["file"])

print("Missing files:", len(missing_files))
```

```python
samples = random.sample(annotations, min(6, len(annotations)))

for item in samples:
    img = cv2.imread(item["file"])

    if img is None:
        continue

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure()
    plt.imshow(img)
    plt.title(f"Label: {item['label']}")
    plt.axis("off")
```

```python
for cls in list(label_counts.keys())[:5]:
    print(f"\nClass: {cls}")

    class_samples = [item for item in annotations if item["label"] == cls]

    if not class_samples:
        continue

    sample = random.choice(class_samples)
    img = cv2.imread(sample["file"])

    if img is None:
        continue

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure()
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")
```

```python
total_images = sum(label_counts.values())
num_classes = len(label_counts)

print("\n📊 Dataset Summary")
print("Total Images:", total_images)
print("Number of Classes:", num_classes)
print("Avg per class:", total_images // num_classes)
```

```python
audio_classes = ["silence", "speech", "noise"]

for cls in audio_classes:
    folder = os.path.join(AUDIO_PATH, cls)

    if os.path.exists(folder):
        count = len(os.listdir(folder))
        print(f"{cls}: {count} files")
```

```python
video_dirs = ["normal_sessions", "cheating_sessions"]

for vd in video_dirs:
    folder = os.path.join(VIDEO_PATH, vd)

    if os.path.exists(folder):
        count = len(os.listdir(folder))
        print(f"{vd}: {count} videos")
```

```python
REPORT_PATH = "../../experiments/results/dataset_summary.json"

os.makedirs(os.path.dirname(REPORT_PATH), exist_ok=True)

summary = {
    "total_images": total_images,
    "num_classes": num_classes,
    "class_distribution": dict(label_counts)
}

with open(REPORT_PATH, "w") as f:
    json.dump(summary, f, indent=4)

print("✅ Dataset summary saved")
```
